```python
def extract_city(location):
    if not isinstance(location, str):
        return None
    parts = re.split(r'\s+in\s+', location, flags=re.IGNORECASE)
    if len(parts) < 2:
        return None
    city = parts[-1].split(',')[0].strip()
    return city 
```

```python
df['city'] = df['location'].apply(extract_city)
df.head(6)
```

```python
df.iloc[10, -2].split(' ·  ·  · ')
```

```python
def numerical_features(misc: str) -> list:
  guests = beds = baths = bedrooms = lux = 0
  
  misc_list = [s for s in misc.split(' ·  ·  · ') if s.strip()]
  patterns = [
    (r'(\d+(?:\.\d+)?)(?:\+)?\s*guests?', 'guests'),   
    (r'(\d+(?:\.\d+)?)\s*(?:bedroom|bedrooms)', 'bedrooms'),
    (r'(\d+(?:\.\d+)?)\s*(?:bed|beds)\b', 'beds'),       
    (r'(\d+(?:\.\d+)?)\s*(?:bath|baths|bathroom|bathrooms)', 'baths'),
    ]
  
  for string in misc_list:
    c=0
    for pattern, var in patterns:
      match = re.search(pattern, string, flags=re.IGNORECASE)
      if match:
        num = float(match.group(1))
        if var == 'guests':
          guests = num
        elif var == 'beds':
          beds = num
        elif var == 'bedrooms':
          bedrooms = num
        elif var == 'baths':
          baths = num
      else:
        c+=1
    if c >=4:
      lux+=1
  return guests, bedrooms, beds, baths, lux     

```

```python
df[['guests', 'bedrooms', 'beds', 'baths', 'luxury_items']] = df['Miscellaneous'].apply( lambda x: pd.Series(numerical_features(x)))
df.head(6)
```

```python
output_df = df.loc[:, ['url', 'description', 'imame_urls', 'city', 'guests', 'bedrooms', 'beds', 'baths', 'luxury_items', 'Price']]
output_df.columns = ['url', 'description', 'image', 'city', 'guests', 'bedrooms', 'beds', 'baths', 'luxury_items', 'Price']
output_df
```
